```python
# collect all paragraphs
paras = []
path = '../data/2_cast_topic_goats.tsv'
qp_ids = []
with open(path, encoding='utf-8') as f:
    for i, line in enumerate(f):
        q_id, para_id, para_text = line.rstrip().split('\t')
        assert '/n' not in para_text
        qp_ids.append((q_id, para_id))
        paras.append(para_text)
print(len(paras))
```

```python
import datetime
import json
import os

squash_path = '/home/svakule/squash-generation'

key = 'trec_cast'
top_p = 0.9
gen_frac = 0.5
spec_frac = 0.8

metadata = {
        "input_text": "\n".join(paras),
        "key": key,
        "timestamp": str(datetime.datetime.now()),
        "settings": {
            "top_p": top_p,
            "gen_frac": gen_frac,
            "spec_frac": spec_frac
        }
    }

os.mkdir("%s/squash/temp/%s" % (squash_path, key))

with open('%s/squash/temp/%s/metadata.json' % (squash_path, key), 'w') as f:
    f.write(json.dumps(metadata))
```
