```python
# GT labels
path = '../data/cast19/train_paragraphs.tsv'
qp_ids = {}
with open(path, encoding='utf-8') as f:
    for i, line in enumerate(f):
        q_id, para_id, para_text = line.rstrip().split('\t')
        if para_id[:5] == 'MARCO':
            q_id = q_id.split('_')[0]
            qp_ids[para_id] = q_id

            
gt_clusters = []
pars = list(qp_ids.keys())
for para_id in pars:
    gt_clusters.append(qp_ids[para_id])
n_clusters = len(gt_clusters)
print("# GT clusters: %d" % n_clusters)
print(gt_clusters)
print(len(gt_clusters))
```

```python
# todo load clusters prodcued by the model
import numpy as np
random_clusters = []
np.random.shuffle(pars)
for para_id in pars:
    random_clusters.append(qp_ids[para_id])
print(random_clusters)
clusters = random_clusters
print(len(clusters))
```

```python
# evaluate clustering
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics.cluster import normalized_mutual_info_score, adjusted_mutual_info_score, fowlkes_mallows_score

# print(gt_clusters)
print(adjusted_rand_score(gt_clusters, clusters))
print(normalized_mutual_info_score(gt_clusters, clusters))
print(adjusted_mutual_info_score(gt_clusters, clusters))
print(fowlkes_mallows_score(gt_clusters, clusters))
```
